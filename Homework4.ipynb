```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load ="../Pandas-Homework-4/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load ="../Pandas-Homework-4/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name"])
df.head()
```

## creating a list of 1 term - the number of scools

```python
# row count = number of schools
schools_count_row = school_data.shape[0]  

total_schools = [schools_count_row]
total_schools
```

```python
# row count = number of students
students_count_row = student_data.shape[0] 

total_students = [students_count_row]
total_students
```

```python
total_budget = [school_data["budget"].sum()]
total_budget
```

```python
average_math_score = [student_data["math_score"].mean()]
average_math_score
```

```python
#finding the number that passed math
passing_math = student_data.loc[student_data["math_score"] >= 70]
num_passing_math = [passing_math.shape[0]]
num_passing_math
```

```python
math_pass_percent = [num_passing_math[0] / total_students[0]]
math_pass_percent
```

```python
average_reading_score = [student_data["reading_score"].mean()]
average_reading_score
```

```python
#finding the number that passed reading
passing_reading = student_data.loc[student_data["reading_score"] >= 70]
num_passing_reading = [passing_reading.shape[0]]
num_passing_reading
```

```python
reading_pass_percent = [num_passing_reading[0] / total_students[0]]
reading_pass_percent
```

```python
overall_passing_rate = [(average_math_score[0] + average_reading_score[0]) / 2]
overall_passing_rate
```

```python
# creating table for district summary
district_summary_df = pd.DataFrame({
    "Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": math_pass_percent,
    "% Passing Reading": reading_pass_percent,
    "% Overall Passing Rate": overall_passing_rate
})
district_summary_df
```

```python
passmathschools = passing_math[['school_name','math_score']].groupby(['school_name']).count()
passmathschools
```

```python
passmathschools.reset_index(inplace=True)
passmathschools
```

```python
passreadingschools = passing_reading[['school_name','reading_score']].groupby(['school_name']).count()
passreadingschools
```

```python
passreadingschools.reset_index(inplace=True)
passreadingschools
```

```python
# adding columns to dataframef
df["Num Passing Math"] = passmathschools["math_score"]
df["Num Passing Reading"] = passreadingschools["reading_score"]
df["% Passing Math"] = df["Num Passing Math"] / df["size"]*100
df["% Passing Reading"] = df["Num Passing Reading"] / df["size"]*100
df
```

```python
# calculating overall passing rate
df["% Overall Passing Rate"] = (df["% Passing Reading"] + df["% Passing Math"] )/2
df
```

```python
del df["Num Passing Math"]
del df["Num Passing Reading"]
df
```

```python
# arrange by top overall percent rate
# show only top five
final_df = df.sort_values(by='% Overall Passing Rate', ascending = False)
final_df.head(5)
```

```python
# renaming columns

final_df = final_df.rename(columns={'reading_score': 'Average Reading Score', 'math_score' : 'Average Math Score',
                                    'type': 'School Type', 'size': 'Total Students', 'budget': 'Total School Budget'})
top_five = final_df.head(5)
top_five
```
