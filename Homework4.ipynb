```python
#gathering neccessary data to start for averae score by school spending per student dataframe
by_school_spending = pd.merge(student_data, school_data, how="left", on=["school_name"])
by_school_spending = by_school_spending.drop(['Student ID', 'gender', 'grade','School ID', 'type'], axis=1)
by_school_spending["Budget Per Student"] = by_school_spending["budget"] / by_school_spending["size"]
by_school_spending.head()
```

```python
# calculating total students that passed math and setting index to budget per student
passing_math = by_school_spending.loc[by_school_spending["math_score"] >= 70]
passing_math.groupby(["Budget Per Student"]).sum()
```

```python
# calculating total students that passed reading and setting index to budget per student
passing_reading = by_school_spending.loc[by_school_spending["reading_score"] >= 70]
passing_reading.groupby(["Budget Per Student"]).sum()
```

```python
# making bins to gather groups of per student budget ranges
bins = [0, 585, 615, 645, 675]
groups = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# placing ranges into bins
by_school_spending["Spending Categories"] = pd.cut(by_school_spending["Budget Per Student"], bins, labels=groups)
by_school_spending = by_school_spending.groupby("Spending Categories")

# calculating number of students per spending category 
count_school_spending = by_school_spending.count()

# showing mean values of table
by_school_spending = by_school_spending.mean()
by_school_spending
```

```python
# showing number of students per spending category 
count_school_spending
```

```python
# finding number of students that passed math per category
passing_math["Spending Categories"] = pd.cut(passing_math["Budget Per Student"], bins, labels=groups)
passing_math = passing_math.groupby(["Spending Categories"])
passing_math = passing_math.count()
passing_math
```

```python
# finding number of students that passed reading per category
passing_reading["Spending Categories"] = pd.cut(passing_reading["Budget Per Student"], bins, labels=groups)
passing_reading = passing_reading.groupby(["Spending Categories"])
passing_reading = passing_reading.count()
passing_reading
```

```python
# calculating % of students that passed math and reading per cateogry by dividing num passed by total per category
by_school_spending["% Passing Math"] = passing_math["student_name"] / count_school_spending["student_name"]*100
by_school_spending["% Passing Reading"] = passing_reading["student_name"] / count_school_spending["student_name"]*100
by_school_spending["% Overall Passing Rate"] = (by_school_spending["% Passing Math"] 
                                                + by_school_spending["% Passing Reading"])/2
by_school_spending = by_school_spending.drop(['size', 'budget', 'Budget Per Student'], axis=1)
by_school_spending = by_school_spending.rename(columns = {"reading_score": "Average Reading Score", "math_score": "Average Math Score"})
by_school_spending
```

```python
#rearranging columns to place math first
cols = by_school_spending.columns.tolist()
cols = cols[1:2] + cols[:1] + cols[2:] 
by_school_spending = by_school_spending[cols]
by_school_spending
```
